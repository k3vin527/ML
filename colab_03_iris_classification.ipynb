```python
data = '/content/drive/MyDrive/data/iris.csv'
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
```

```python
# seed 값 설정

np.random.seed(42)
tf.random.set_seed(42)
```

```python
#  데이터 입력
pd.read_csv(data, 
            names = ['sepal_llength', 'sepal_width', 'petal_length', ' petal_width', 'species'])
```

```python
df = pd.read_csv(data, 
            names = ['sepal_llength', 'sepal_width', 'petal_length', ' petal_width', 'species'])
```

```python
# 그래프로 확인
sns.pairplot(df, hue='species')
plt.show()
```

```python
df.info()
```

```python
# 데이터 분류
df.values
```

```python
dataset = df.values

x = dataset[:, :-1].astype(float)
y_obj = dataset[:, -1] # y값이 문자열(object)
```

```python
# 문자열 >> 숫자 변환 (label_encoder 사용)

le = LabelEncoder()
le.fit(y_obj)
y = le.transform(y_obj)
```

```python
y
```

```python
y_encoded = tf.keras.utils.to_categorical(y)
# 활성화 함수 적용하기 위해서 y 값을 0,1로 변경 (one_hot encoding)
```

```python
# 모델 설정
model = Sequential()
model.add(Dense(16, input_dim= 4, activation='relu'))
model.add(Dense(3, activation = 'softmax'))
```

```python
# 모델 컴파일
model.compile(loss = 'categorical_crossentropy',   # 범주형 크로스엔트로피 사용 (일반적)
              optimizer = 'adam',                   # 최적화 함수  :adam
              metrics = ['accuracy'])               # 평가기준 : accuracy(정확도)
```

```python
# 모델 실행
model.fit(x, y_encoded, epochs = 50, batch_size = 1)
```

```python
# 결과 출력
print('\n Accuracy : %.4f'%(model.evaluate(x,y_encoded)[1]))
```
